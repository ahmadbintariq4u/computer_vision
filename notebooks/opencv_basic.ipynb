## 10 operations of OpenCv

```python
import numpy as np
import pandas as pd
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image=cv.imread('image.png')
```

### 1. Merging & Spliting

```python
b,g,r=cv.split(image)   # spliting
image=cv.merge([r,g,b]) # merging
```

### 2. Resizing

```python
# Matplot grids.
# Setting DPI for increasing size.
plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 150

f = plt.figure()
f.add_subplot(1,2, 1)
plt.xlabel('Input')
plt.imshow(image)

plt. subplots_adjust(right=1.5)  # adjust size between images
f.add_subplot(1,2,2)
plt.xlabel('Output')

# resizing.
output=cv.resize(image,(500,500))
plt.imshow(output)
```


### 3. Rotation

```python
width,height=image.shape[:2]
center = (width/2, height/2)
rotate_matrix = cv.getRotationMatrix2D(center=center, angle=45, scale=1)
rotated_image = cv.warpAffine(src=image, M=rotate_matrix, dsize=(width, height))

f = plt.figure()
f.add_subplot(1,2, 1)
plt.xlabel('Input')
plt.imshow(image)

plt. subplots_adjust(right=1.5)  # adjust size between images
f.add_subplot(1,2,2)
plt.xlabel('Output')
plt.imshow(rotated_image)
```

### 4. Translation

```python

translation_matrix = np.array([
    [1, 0, 1],
    [0, 1, 1]
], dtype=np.float32)

translated_image = cv.warpAffine(src=image, M=translation_matrix, dsize=(width, height))

f = plt.figure()
f.add_subplot(1,2, 1)
plt.xlabel('Input')
plt.imshow(image)

plt. subplots_adjust(right=1.5)  # adjust size between images
f.add_subplot(1,2,2)
plt.xlabel('Output')
plt.imshow(translated_image)

```

### 5. Cropping

```python
orignial_image=cv.cvtColor(image,cv.COLOR_RGB2GRAY)
copied_array=orignial_image[300:700,300:700]

copied_array=cv.copyMakeBorder(copied_array,top=15,bottom=15,left=15,right=15,borderType=cv.BORDER_CONSTANT)

f = plt.figure()
f.add_subplot(1,2, 1)
plt.xlabel('Input')
plt.imshow(orignial_image,cmap='gray')

orignial_image[700:1130,700:1130]=copied_array
# orignial_image=copied_array
plt. subplots_adjust(right=1.5)  # adjust size between images
f.add_subplot(1,2,2)
plt.xlabel('Output')
plt.imshow(orignial_image,cmap='gray')
```

### 6. Mirroring

```python
# Flipping
orignial_image=cv.cvtColor(image,cv.COLOR_RGB2GRAY)
flipped=cv.flip(orignial_image,0)

f = plt.figure()
f.add_subplot(1,3, 1)
plt.xlabel('Input Mirroing')
plt.imshow(orignial_image,cmap='gray')

plt.subplots_adjust(right=1.5)  # adjust size between images
f.add_subplot(1,3,2)
plt.xlabel('Output Flipping')
plt.imshow(flipped,cmap='gray')


mirrored=cv.flip(orignial_image,1)
f.add_subplot(1,3, 3)
plt.xlabel('Output Mirroring')
plt.imshow(orignial_image,cmap='gray')
```

### 7. Image Negatives

```python
orignial_image=cv.cvtColor(image,cv.COLOR_RGB2GRAY)

f=plt.figure()
f.add_subplot(1,2,1)
plt.xlabel('Input Image')
plt.imshow(orignial_image,cmap='gray')
plt.subplots_adjust(right=1.5)  # adjust size between images

negative=abs(255-orignial_image)
f.add_subplot(1,2,2)
plt.xlabel('Output Image')
plt.imshow(negative,cmap='gray')
```

### 8. Blur using Box Filter

```python
sharp_image=cv.imread('sharp.jpg',0)

f=plt.figure()
f.add_subplot(1,2,1)
plt.xlabel('Input Sharp Image')
plt.imshow(sharp_image,cmap="gray")
plt.subplots_adjust(right=1.5)

dst=cv.blur(sharp_image,(5,5))

f.add_subplot(1,2,2)
plt.xlabel("Output Smooth Image")
plt.imshow(dst,cmap="gray")


```

### 9. Convolution : Linear Filtering

```python
first=cv.imread('logo.png',0)

f=plt.figure()
f.add_subplot(1,2,1)
plt.xlabel('Input Sharp Image')
plt.imshow(first,cmap="gray")
plt.subplots_adjust(right=1.5)


kernel = np.ones((3,3),np.float32)/9
dst = cv.filter2D(first,-10,kernel)

f.add_subplot(1,2,2)
plt.xlabel("Output Smooth Image")
plt.imshow(dst,cmap="gray")

```

### 10. Median Blurring

```python
first=cv.imread('logo.png',0)

f=plt.figure()
f.add_subplot(1,2,1)
plt.xlabel('Input Sharp Image')
plt.imshow(first,cmap="gray")
plt.subplots_adjust(right=1.5)

median = cv.medianBlur(first,5)

f.add_subplot(1,2,2)
plt.xlabel("Output Smooth Image")
plt.imshow(median,cmap="gray")
```
